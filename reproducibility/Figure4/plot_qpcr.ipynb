```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import mplscience
import pandas as pd
import seaborn as sns
```

```python
qpcr = pd.read_csv("./data/kpspatial_hypoxia_qpcr.txt", sep='\t', index_col = 0)
qpcr.head(5)
```

```python
qpcr.columns
```

```python
cols = ['FC (LT-HT)', 'HG vs HT', 'FC (LG-HG)', 'LG-HG / LT-HT']
qpcr_subset = qpcr[cols]
qpcr_subset
```

```python
with mplscience.style_context():
    # fig = plt.figure(figsize=(5, 3))
    sns.heatmap(qpcr_subset, cmap='RdBu_r', center=1, robust=True, linewidths=0.5, linecolor='black', vmax=3)
    plt.xticks([0, 1, 2, 3], ['LT vs HT', 'HG vs HT', 'LG vs HG', '(LG vs HG) / (LT vs LG)'], rotation=30)
    plt.title("dCq FCs") 
    plt.show()
```
